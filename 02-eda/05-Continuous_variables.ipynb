````python
df_sample['date_'] = df_sample['date'].apply(lambda x : pd.to_datetime(x))
````

````python
df_sample[['date_', 'date']].head()
````

````python
df_sample_sorted = df_sample.sort_values(by='date_')
````

````python
df_sample_sorted.head()
````

````python
df_sample_sorted.date_.min(), df_sample_sorted.date_.max()
````

````python
fig = sns.relplot(data=df_sample_sorted, x='date', y="item_price", kind='line', height = 5, aspect=3);
fig.set_xticklabels(rotation=30);
````

````python
fig = sns.relplot(data=df_sample_sorted, x='date_', y="item_price", kind='line', height = 5, aspect=3);
fig.set_xticklabels(rotation=30);
````
