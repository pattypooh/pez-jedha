# What attributes the participants search for?

```python
searched_attributes = ['iid', "attr1_1", "sinc1_1", "intel1_1", "fun1_1", "amb1_1", "shar1_1" , "gender"]
```

```python
preferences_df = data_df.groupby(searched_attributes, as_index=False).sum()[searched_attributes]
```

```python
pref_mean = preferences_df.groupby("gender", as_index=False).mean()
pref_mean
```

```python
pref_cat = ["attr1_1", "sinc1_1", "intel1_1", "fun1_1", "amb1_1", "shar1_1"]
male_vals = pref_mean[(pref_mean.gender ==1)][pref_cat]
```

```python
new_pref_mean = pd.melt(pref_mean, id_vars="gender", var_name="attribute", value_name="means")
```

```python
new_pref_mean.drop(index=[0,1], inplace=True)
```

```python
new_pref_mean
```
